Input 6114526464654131546783427384239

Input 611452646-901413154-683342738-456457821

En python se puede alinear texto a la izquierda, centro o derecha.



**.ljust(width)**

**.center(width)**


> width = 20

**.rjust(width)**



> width = 20
>
> print 'Palabra'.rjust(width,'-')
>
> ----------Palabra

Task

Dada esta breve introduccion de como usar estos metodos.
Reutilizar el codigo y generar la letra H con el formato de Salida siguiente. Usar solo dos variables con los valores:

thickness = 5
letter = 'H'

Ouput:

H.JPG

```python
l = ''
thickness = 5
letter = 'H'

for i in range(thickness):
    print ((letter*(i+1)).ljust(thickness))
for i in range(thickness):
    print ((letter*thickness).center(thickness) + (letter*thickness).center(thickness*5))
for i in range((thickness+1)//2):
    print ((letter*thickness*4).center(thickness*4))
for i in range(thickness):
    print ((letter*thickness).center(thickness) + (letter*thickness).center(thickness*5))
for i in range(thickness):
    print ((letter*(thickness-i)).rjust(thickness*4))
```
